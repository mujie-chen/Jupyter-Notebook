# Plotting with Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme()

biomarkers = pd.read_csv('seaborn_demo.csv', usecols=lambda column: "Unnamed" not in column)
print(biomarkers.head())
```

```python
from numpy import random
x = random.normal(loc=1, scale=2, size=(2,3)) # loc = Mean; scale = SD

print(x)
```

## Single Variable Frequency

```python
sns.displot(biomarkers, x="MUNE", bins=30)
```

```python
from numpy import random

sns.displot(random.normal(size=50000), bins=1000)

plt.show()
```

## Two Variable Frequency

```python
sns.relplot(data=biomarkers, x="LDH (U/L - Serum)", y="Asparate Aminotransferase (U/L- Serum)", hue="Gender", palette="Reds")
```
